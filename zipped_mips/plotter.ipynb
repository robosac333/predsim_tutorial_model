```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_data(folder_path):
    plot_store = []

    fig, ax = plt.subplots(len([f for f in os.listdir(folder_path) if f.endswith(".csv")]), figsize=(20, 20))
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            # Skip the first row and use the second row as headers
            data = pd.read_csv(folder_path + '\\' + filename, header=1, skiprows=[0])
            plot_store.append(data)
    
    for idx, data in enumerate(plot_store):
        ax[idx].plot(data.iloc[:, 0], data.iloc[:, 1])  # Use iloc to access columns by index
        ax[idx].set_xlabel('Time')
        ax[idx].set_ylabel('Actuator Force')
        ax[idx].set_title(f'Actuator Force vs Time ({os.path.splitext(os.listdir(folder_path)[idx])[0]})')
    
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    folder_path = r"C:\Users\sachi\predsim_tutorial_model\zipped_mips\9-18-Force_Trials"
    # folder_path = r"C:\Users\sjd3333\Desktop\MIPS_data\visit26trials"
    plot_data(folder_path)
```
